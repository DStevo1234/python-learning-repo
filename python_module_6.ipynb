```python
# Fucntions
# A way to wirte bits of code to use repeatedly whilst changing inputs

def print_variable(x):
    phrase = f'My name is {x}'
    print(phrase)

print_variable("Dan")

def sum_xy(x, y):
    z = x + y
    print(z)

sum_xy(x=10, y=5)

# Can pass by variable by locaiton (x,y) = (5,10) -- order matters
# Cam pass variable by name (x,y) = (y=10, x=5) -- order doesn't matter

def minus_xy(x, y):
    w = x - y
    return w

answer = minus_xy(15, 5)

print(answer)

# variables defined in funcitons only exist inside function (Scope)
# variables defined outisde functions can be used in funcitons

def maths_xy(x, y):
    minus = x - y
    plus = x + y
    return minus, plus

# print(maths_xy(150,80))

# print(maths_xy(150,80)[0]) # takes first value in tuple

# print(maths_xy(150,80)[1]) # takes second value in tuple

# x_minus_y, x_plus_y = maths_xy(150,80)
# print(x_minus_y)
# print(x_plus_y)

def moremaths_xy(x, y):
    times = x * y
    divide = x / y
    return times, divide

x_times_y, x_divide_y = moremaths_xy(100,20)
print(x_times_y)
print(int(x_divide_y)) # useful for ages (always rounds down)

# Nesting functions

# def function_1(x, y):
  #  return f'function 1 says x and y are {x}, and {y}'

# def function_2(x, y):
   # return f'fucntion 2 says x and y are {x}, and {y}'

# def nested_func(j, k):
   # var_1 = function_1(j, k)
   # var_2 = function_2(j, k)
   # return var_1, var_2

# print(nested_func("Hi","Hello"))

def function_forename(forename):
    return f'Hello, my forename is {forename}'

def function_surname(surname):
    return f'My surname is {surname}'

def func_name(fname, sname):
    fore_1 = function_forename(fname)
    sur_1 = function_surname(sname)
    return fore_1, sur_1

names = func_name("Dan","Stevens")
print(names)

import pandas as pd

url = 'https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv'

df = pd.read_csv(url)

# function to work out if percentage change monthly is above or below average and add to dataframe

mean_pct_change = df['Percentage change (monthly) All property types'].mean()

# print(mean_pct_change)

def change_checker(pct_change):
    if pct_change > mean_pct_change:
        return 'Above average'
    if pct_change < mean_pct_change:
        return 'Below average'
    else:
        return 'Exactly average'
    
df['Average change percentage check'] = df['Percentage change (monthly) All property types'].apply(change_checker)

mean_house_prices = df['Average price All property types'].mean()

def hp_checker(house_price):
    if house_price > mean_house_prices:
        return 'Above house average'
    if house_price < mean_house_prices:
        return 'Below house average'
    else:
        return 'Exactly house average'
    
df['Average house price check'] = df['Average price All property types'].apply(hp_checker)

df
```
