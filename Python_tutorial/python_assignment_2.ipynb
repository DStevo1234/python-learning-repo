```python
# Creation of dataframe

import pandas as pd

HR = pd.DataFrame([
    {'EmployeeID':'12',
     'Age':25,
     'Salary':30000,
     'Job Title':'MIO'},
     {'EmployeeID':'20',      
      'Age':35,
    'Salary':37500,
    'Job Title':'Analyst'},
     {'EmployeeID':'09',      
      'Age':40,
    'Salary':50000,
    'Job Title':'Techincal Analyst'},
     {'EmployeeID':'94',      
      'Age':45,
    'Salary':67500,
    'Job Title':'Manager'},
     {'EmployeeID':'48',      
      'Age':50,
    'Salary':90000,
    'Job Title':'Director'}
 ])

# Min, Mean, Max  and Sum calculations for age and salary

Min_Salary = HR['Salary'].min()
Max_Salary = HR['Salary'].max()
Mean_Salary = HR['Salary'].mean()

Min_Age = HR['Age'].min()
Max_Age = HR['Age'].max()
Mean_Age = HR['Age'].mean()

Sums = HR.sum(axis=0,skipna=True, numeric_only=True, min_count=0)

print('Min Salary =',Min_Salary)
print('Max Salary =',Max_Salary)
print('Mean Salary =',Mean_Salary)

print('Min Age =',Min_Age)
print('Max Age =',Max_Age)
print('Mean Age =',Mean_Age)

print(Sums)


```

```python
# Creation of dataframe

import pandas as pd

HR = pd.DataFrame([
    {'EmployeeID':'12',
     'Age':25,
     'Salary':30000,
     'Job Title':'MIO'},
     {'EmployeeID':'20',      
      'Age':35,
    'Salary':37500,
    'Job Title':'Analyst'},
     {'EmployeeID':'09',      
      'Age':40,
    'Salary':50000,
    'Job Title':'Techincal Analyst'},
     {'EmployeeID':'94',      
      'Age':45,
    'Salary':67500,
    'Job Title':'Manager'},
     {'EmployeeID':'48',      
      'Age':50,
    'Salary':90000,
    'Job Title':'Director'}
 ])

# Add slice to identify all salaries above the mean

Mean_Salary = HR['Salary'].mean()

Above_mean = HR.loc[HR['Salary'] > Mean_Salary]

print(Above_mean)
```
